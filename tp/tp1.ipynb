## `Mesure de performance`
(apprentissage supervise)

### 1. exactitude (accuracy)

l'exactitude est une meseure de performance indiquant a quel point le prediction d'un model est celui qu'il doit etre, ( en gros defini a quel point le modele est exact dans ces previsions)

```python
# exemple 1

TP, TN, FP, FN = 42, 32, 8, 18
exactitude = (TP + TN)/(TP + TN + FP + FN)
print(exactitude)
```

il faut noter qu'il ne garantit par l'obtention d'un model car un model peut etre exact , mais pas precis dans les predictions d'un type de classe precis

```python
# exemple 2

TP, TN, FP, FN = 0, 95, 5, 0
exactitude = (TP + TN)/(TP + TN + FP + FN)
print(exactitude)
```

### 2. precision

differente de l'exactitude, elle permet de decrire a quel point le modele donne les bonnes predictions

```python
# exemple 1

TP, TN, FP, FN = 0, 95, 5, 0
precision = ( (TP /(TP + FP)) + (TN /(TN + FN)) )/2
print(precision)
```

```python
# exemple 2

TP, TN, FP, FN = 114, 12, 14, 0
precision = ( (TP /(TP + FP)) + (TN /(TN + FN)) )/2
print(precision)
```

### 3. rappel

encore appele sensibilite, decrit a quel point le model est sensible et identifie les classes

```python
# exemple 2

TP, TN, FP, FN = 114, 12, 14, 0
rappel = ( (TP /(TP + FN)) + (TN /(TN + FP)) )/2
print(rappel)
```

### 4. fmesure 
permet d'avoir un ratio de la presicion et du rappel du modele

```python
fmesure = 2*precision*rappel/(precision + rappel )
print(fmesure)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
iris = load_iris()
X = []
for ind_class in range(3):
    X.append([[], [], []])
    for i in range(len(iris.data)):
        if iris.target[i] == ind_class:
            X[ind_class][0].append(iris.data[i][0])
            X[ind_class][1].append(iris.data[i][1])
            X[ind_class][2].append(sum(iris.data[i][2:]))
colours = ("r", "g", "y")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for ind_class in range(3):
    ax.scatter(X[ind_class][0], X[ind_class][1], X[ind_class][2], c=colours[ind_class])
plt.show()
```

#### decoupage de donnees en donnees d'entrainement et de test

1. holdout (decoupage aleatoire)

```python
# decoupage avec numpy de base 
indices = np.random.permutation(len(iris.data))

n_data_train = 50
data_train = iris.data[indices[:n_data_train],]
data_test = iris.data[indices[-n_data_train:],]
label_train = iris.target[indices[:n_data_train],]
label_test = iris.target[indices[-n_data_train:],]

print("** donnees d'entrainnement (9 premiers):\n", data_train[:9,], "\n cibles :", label_train[:9])
print("** donnees de test (9 premiers):\n", data_test[:9,], "\n cibles :", label_test[:9,])
```

```python
## avec sklearn 

from sklearn.model_selection import train_test_split
iris = load_iris()
data, labels = iris.data, iris.target
res= train_test_split(
    data, labels, train_size = 0.8, test_size= 0.2, random_state=42
)
data_train, data_test, target_train, target_test = res
print("** donnees d'entrainnement (9 premiers):\n", data_train[:9,], "\n cibles :", label_train[:9])
print("** donnees de test (9 premiers):\n", data_test[:9,], "\n cibles :", label_test[:9,])
```

2. decoupage aleatoire en homogenisant 

On homogenise les pourcentages d'apparition des cibles en considerant la dataset de depart

```python
# Verification des pourcentages pour le cas precedent 
# np.bitcount permet d'obtenir un tableau d'occurences des elements 
# du tableau passer en paramettre
print('toutes la dataset :', np.bincount(labels) / float(len(labels)) * 100.0)
print('entrainement:', np.bincount(label_train) / float(len(label_train)) * 100.0)
print('test:', np.bincount(label_test) / float(len(label_test)) * 100.0)
```

```python
# realisation du decoupage homogene avec sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
data, labels = iris.data, iris.target
res = train_test_split(data, labels,
            train_size=0.8,
            test_size=0.2,
            random_state=42,
            stratify=labels)
train_data, test_data, label_train, label_test = res
print('toutes la dataset :', np.bincount(labels) / float(len(labels)) * 100.0)
print('entrainement:', np.bincount(label_train) / float(len(label_train)) * 100.0)
print('test:', np.bincount(label_test) / float(len(label_test)) * 100.0)
```
